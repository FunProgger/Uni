```python
 #Biomeng261_L1_Part1.py
import matplotlib.pyplot as plt
import numpy as np
from scipy.integrate import solve_ivp   
```

```python
file_path = 'Biomeng261_L1_Part2_Data.csv'

conc_s = [] 
rxnVel_v = []

with open(file_path, 'r') as file:
        file.readline()
        for line in file:
            values = line.strip().split(',')
            if len(values) >= 2:
                conc_s.append(float(values[0]))
                rxnVel_v.append(float(values[1]))

plt.plot(rxnVel_v, conc_s,'ko')
plt.xlabel('[S]')
plt.ylabel('v')
plt.title('Hill Curve')
```

```python
def hill(S, n):
    Kd = 60
    Vmax = 30
    v_hill = Vmax * np.power(S, n) / (Kd + np.power(S, n))
    return v_hill

hill(conc_s, 2)
```

```python
for i in range(2, 6):
    plt.plot(hill(conc_s,i), conc_s, label = f'n = {i}')
    plt.xlabel('[S]')
    plt.ylabel('v')
    plt.title('Hill Curve')

plt.plot(rxnVel_v, conc_s, linestyle = '--', label = 'measured')

# n = 2.6
# plt.plot(hill(conc_s,n), conc_s, label = f'n = {n}')

plt.legend()
    
```
